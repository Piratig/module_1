# Functions (Функции)


```python
def visualise_model(Y_test, probs, y_pred):
    
    #### ROC AUC
    fpr, tpr, threshold = roc_curve(Y_test, probs)
    roc_auc = roc_auc_score(Y_test, probs)

    fig, ax = plt.subplots(figsize=(10,5))
    plt.plot([0, 1], label='Baseline', linestyle='--')
    plt.plot(fpr, tpr, label = 'Regression')
    ax.set_title('ROC AUC = %0.3f'%roc_auc)
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.legend(loc='lower right')
    
    #### confusion matrix
    fig, ax = plt.subplots(figsize=(10,5))
    sns.set_context(context='paper', font_scale=2, rc=None)
    group_names = ['True Pos', 'False Pos', 'False Neg', 'True Neg']
    group_counts = ['{0:0.0f}'.format(value) for value in
                    confusion_matrix(Y_test, y_pred).flatten()]
    labels = [f'{v1}\n{v2}' for v1, v2 in
              zip(group_names, group_counts)]
    labels = np.asarray(labels).reshape(2, 2)
    ax = sns.heatmap(confusion_matrix(Y_test, y_pred),
                     annot=labels, fmt='', cmap='Greens')
    ax.set(xlabel='predicted', ylabel='real',title = 'Confusion matrix')
    plt.show()
```

```python
# чиловая группа
num_cols = ['age', 'decline_app_cnt', 'bki_request_cnt',
            'income', 'score_bki', 'region_rating']

# категории
cat_cols = ['education', 'work_address', 'home_address', 'sna', 'first_time']
```

```python
### Переведу в категориальный признак месяцев ###

data['app_date'] = pd.to_datetime(data.app_date)

data['month'] = data.app_date.apply(lambda x: x.month)
cat_cols.append('month')
```

```python
cat_cols_dummies = pd.get_dummies(data[cat_cols])
```

```python
imp_cat = Series(mutual_info_classif(data[bin_cols].drop('train', axis=1),
                                data['default'],
                                discrete_features=True), index=bin_cols[:-1])
imp_cat.sort_values(inplace=True)
imp_cat.plot(kind='barh')
```

```python
# численные
imp_cat = Series(mutual_info_classif(data[num_cols],
                                data['default']), index=num_cols)
imp_cat.sort_values(inplace=True)
imp_cat.plot(kind='barh')
```

```python
# категориальные
imp_cat = Series(mutual_info_classif(cat_cols_dummies,
                                data['default']), index=cat_cols_dummies.columns)
imp_cat.sort_values(inplace=True)
imp_cat.plot(kind='barh')
```

```python
# обработка категориальных переменных
data = pd.get_dummies(data, columns = cat_cols)
```

```python
# Логарифмирование "хвостатых переменных" среди численных
num_cols_tails = ['income', 'decline_app_cnt', 'bki_request_cnt']

for col in num_cols_tails:
    data[col] = np.log(data[col] + 1)
```

```python
# Стандартизация численных переменных
data[num_cols] = pd.DataFrame(StandardScaler().fit_transform(data[num_cols]),
                              columns=num_cols)
```

```python
# Разделим выборку обратно на тест и трейн
train = data.query('train == 1').drop(['train'], axis=1)
test = data.query('train == 0').drop(['train'], axis=1)
X_test = test.drop(columns = ['default', 'client_id','app_date'])
```

```python
RANDOM_SEED = 42
X_data_train = train.drop(columns = ['default', 'client_id','app_date'])
Y_data_train = train['default']
```

# Models

## Basic - no balance

```python
X_train, X_valid, Y_train, Y_valid = train_test_split(X_data_train, Y_data_train,
                                               random_state=RANDOM_SEED,
                                               test_size=0.2)


model = LogisticRegression()
model.fit(X_train, Y_train)
probs = model.predict_proba(X_valid)
y_pred = model.predict(X_valid)
probs = probs[:,1]

print('F1 = ' + str(f1_score(Y_valid, y_pred)))
visualise_model(Y_valid, probs, y_pred) # вывод ROC AUC и confusion matrix
```

## Under sample

```python
# Выровним количество дефолтов с количеством их отсутствия.
# Для этого случайно урежем выборку клиентов без дефолта.
under_train = pd.concat([train[train['default'] == 0].sample(
    len(train[train['default'] == 1]), random_state=RANDOM_SEED),
                         train[train['default'] == 1]])

X_under_train = under_train.drop(columns=['default', 'client_id', 'app_date'])
Y_under_train = under_train['default']

X_train, X_valid, Y_train, Y_valid = train_test_split(
    X_under_train, Y_under_train, random_state=RANDOM_SEED, test_size=0.2)

model = LogisticRegression()
model.fit(X_train, Y_train)
y_pred = model.predict(X_valid)
probs = model.predict_proba(X_valid)
probs = probs[:, 1]

print('F1 = ' + str(f1_score(Y_valid, y_pred)))
visualise_model(Y_valid, probs, y_pred)
```

## SMOTE

```python
sm = SMOTE(sampling_strategy='minority', random_state=RANDOM_SEED)
X_sm, Y_sm = sm.fit_resample(X_data_train, Y_data_train)

X_train, X_valid, Y_train, Y_valid = train_test_split(
    X_sm, Y_sm, random_state=RANDOM_SEED, test_size=0.2)

model = LogisticRegression()
model.fit(X_train, Y_train)
y_pred = model.predict(X_valid)
probs = model.predict_proba(X_valid)
probs = probs[:, 1]

print('F1 = ' + str(f1_score(Y_valid, y_pred)))
visualise_model(Y_valid, probs, y_pred)
```

## Random over sample

```python
ros = RandomOverSampler(sampling_strategy='minority', random_state=RANDOM_SEED)
X_ros, Y_ros = ros.fit_resample(X_data_train, Y_data_train)

X_train, X_valid, Y_train, Y_valid = train_test_split(
    X_ros, Y_ros, random_state=RANDOM_SEED, test_size=0.2)

model = LogisticRegression()
model.fit(X_train, Y_train)
y_pred = model.predict(X_valid)
probs = model.predict_proba(X_valid)
probs = probs[:, 1]

print('F1 = ' + str(f1_score(Y_valid, y_pred)))
visualise_model(Y_valid, probs, y_pred)
```

## Borderline SMOTE

```python
smBL = BorderlineSMOTE(sampling_strategy='minority', random_state=RANDOM_SEED)
X_smBL, Y_smBL = smBL.fit_resample(X_data_train, Y_data_train)

X_train, X_valid, Y_train, Y_valid = train_test_split(
    X_smBL, Y_smBL, random_state=RANDOM_SEED, test_size=0.2)

model = LogisticRegression()
model.fit(X_train, Y_train)
y_pred = model.predict(X_valid)
probs = model.predict_proba(X_valid)
probs = probs[:, 1]

print('F1 = ' + str(f1_score(Y_valid, y_pred)))
visualise_model(Y_valid, probs, y_pred)
```

# Application of the model

```python
f1_max = 0
th = 0
pred_proba_df = pd.DataFrame(model.predict_proba(X_valid))
threshold_list = [x/100 for x in range(5, 100, 5)]
for i in threshold_list:
    y_pred = pred_proba_df.applymap(lambda x: 1 if x>i else 0).values[:, 1]
    test_f1_score = f1_score(Y_valid, y_pred)
    if test_f1_score > f1_max:
        f1_max = test_f1_score
        th = i
print(f'Max F1 score is {f1_max} with {th} threshold')
```

```python
f1_max_fine = 0
th_fine = 0
pred_proba_df = pd.DataFrame(model.predict_proba(X_valid))
threshold_list = [x/100 for x in range(int(th*100 - 10), int(th*100 + 10))]
for i in threshold_list:
    y_pred = pred_proba_df.applymap(lambda x: 1 if x>i else 0).values[:, 1]
    test_f1_score = f1_score(Y_valid, y_pred)
    if test_f1_score > f1_max_fine:
        f1_max_fine = test_f1_score
        th_fine = i
print(f'Max F1 score is {f1_max_fine} with {th_fine} threshold')
```

# Libraries (Библиотеки)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import combinations
from scipy.stats import ttest_ind
```

# First data review (Первичный отсмотр данных)

```python
pd.set_option('display.max_rows', 50) # показывать больше строк
pd.set_option('display.max_columns', 50) # показывать больше колонок

# Выгрузка данных в DataFrame из .csv файла
df_stud_math = pd.read_csv('stud_math.csv')
# Замена названия одново из столбцов для добной работы
df_stud_math.columns = ['studytime_granular' if x=='studytime, granular' else x for x in df_stud_math.columns]

display(df_stud_math.head(10))
df_stud_math.info() 
```

# Functions (Функции)

```python
# Обработка числовых данных в столбце
def number_processing(col_name):  
    print('{},'.format(df_stud_math[col_name].describe()))
    df_stud_math[col_name].hist()
    
# Заполение пропсков сохраняя процентное соотношение
def fill_gaps_by_ratio(col_name):
    df_pr = pd.DataFrame(df_stud_math[col_name].value_counts())
    a = df_stud_math[col_name].isnull().sum()
    b = df_pr[col_name].sum()
    df_pr['percent'] = df_pr[col_name].apply(lambda x: x*100/b)
    df_pr['fill_with_percent'] = df_pr['percent'].apply(lambda x: x*a/100).round().astype(int)
    df_pr = df_pr.loc[df_pr['fill_with_percent'] != 0]
    for n in range(df_pr.shape[0]):
        df_stud_math[col_name] = df_stud_math[col_name].fillna(value=df_pr.index[n], limit=df_pr.iloc[n]['fill_with_percent'].astype(int))
    col_info(col_name)
    
# Вывод инфогрмаии о данных в столбце
def col_info(col_name):
    print('Количество пропусков: {}'.format(df_stud_math[col_name].isnull().sum()))
    display(pd.DataFrame(df_stud_math[col_name].value_counts()))
    print("Значений, встретившихся в столбце более 10 раз:", (df_stud_math[col_name].value_counts()>10).sum())
    df_stud_math.loc[:, [col_name]].info()

# Фильтрация выбросов формулой интерквартильного расстояния (межквартильного размаха)
def iqr_calculation(col_name):
    IQR = df_stud_math[col_name].quantile(0.75) - df_stud_math[col_name].quantile(0.25)
    perc25 = df_stud_math[col_name].quantile(0.25)
    perc75 = df_stud_math[col_name].quantile(0.75)
    print('25-й перцентиль: {},'.format(perc25), '75-й перцентиль: {},'.format(perc75), "IQR: {}, ".format(IQR), "Границы выбросов: [{f}, {l}].".format(f=perc25 - 1.5*IQR, l=perc75 + 1.5*IQR))
 
    df_stud_math[col_name].loc[df_stud_math[col_name].between(perc25 - 1.5*IQR, perc75 + 1.5*IQR)].hist(bins = 10, range = (0, 10), label = 'IQR')
 
    plt.legend()

# Построения графиков плотность распределения переменных
def get_boxplot(column):
    fig, ax = plt.subplots(figsize = (10, 4))
    sns.boxplot(x=column, y='score', 
                data=df_stud_math.loc[df_stud_math.loc[:, column].isin(df_stud_math.loc[:, column].value_counts().index[:])], 
                ax=ax)
    plt.xticks(rotation=45)
    ax.set_title('Boxplot for ' + column)
    plt.show()

# Оценок по номинативным признакам, с помощью теста Стьюдента.
def get_stat_dif(column):
    cols = df_stud_math.loc[:, column].value_counts().index[:]
    combinations_all = list(combinations(cols, 2))
    for comb in combinations_all:
        if ttest_ind(df_stud_math.loc[df_stud_math.loc[:, column] == comb[0], 'score'], 
                        df_stud_math.loc[df_stud_math.loc[:, column] == comb[1], 'score']).pvalue \
            <= 0.05/len(combinations_all): # Учли поправку Бонферони
            print('Найдены статистически значимые различия для колонки', column)
        break
```
